```python
import pandas as pd
```

```python
df = pd.read_csv("proyecto_modular.csv")
df.head()
```

```python
inputs = df.drop('Instrumento Id',axis='columns')
```

```python
target = df['Instrumento Id']
```

```python
from sklearn.preprocessing import LabelEncoder
le_time = LabelEncoder()
le_risk = LabelEncoder()
```

```python
inputs['time'] = le_time.fit_transform(inputs['Tiempo'])
inputs['risk'] = le_risk.fit_transform(inputs['Riesgo'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['Id', 'Tiempo','Instrumento nombre', 'Riesgo', 'Instrumento rendimiento'],axis='columns')
```

```python
inputs_n
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n.values, target.values)
```

```python
model.score(inputs_n.values,target)
```

```python
model.predict([[0, 0]])
```

```python
import joblib
joblib.dump(model, 'model.pkl', compress=9)
```

```python
model_clone = joblib.load('model.pkl')
model_clone.predict([[0, 0]])
```
